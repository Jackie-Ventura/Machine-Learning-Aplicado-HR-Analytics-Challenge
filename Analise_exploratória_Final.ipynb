```python
#import pandas as pd

def detectar_outliers_em_todas(df):
    df_numerico = df.select_dtypes(include='number')
    resultado = {}

    for col in df_numerico.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lim_inf = Q1 - 1.5 * IQR
        lim_sup = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lim_inf) | (df[col] > lim_sup)][col]
        resultado[col] = {
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Limite Inferior': lim_inf,
            'Limite Superior': lim_sup,
            'NÂº Outliers': outliers.count(),
            '% Outliers': round(outliers.count() / len(df) * 100, 2)
        }

    return pd.DataFrame(resultado).T.sort_values(by='% Outliers', ascending=False)

# Aplicar ao seu DataFrame
outliers_df = detectar_outliers_em_todas(df)
outliers_df
```

```python
# MÃ©dia salarial por cargo e razÃ£o salarial
df['AvgSalaryByRole'] = df.groupby('JobRole')['MonthlyIncome'].transform('mean')
df['SalaryRatio'] = df['MonthlyIncome'] / df['AvgSalaryByRole']
# Risco de estagnaÃ§Ã£o
df['StagnationRisk'] = df['YearsSinceLastPromotion'] * df['JobLevel']
# Treinamento por ano na empresa
df['TrainPerYear'] = df['TrainingTimesLastYear'] / (df['YearsAtCompany'] + 1)
# ProporÃ§Ã£o de anos na funÃ§Ã£o
df['YearsPerRoleRatio'] = df['YearsInCurrentRole'] / (df['TotalWorkingYears'] + 1)
# Tempo com o gerente atual
df['ManagerTenure'] = df['YearsWithCurrManager']
# Ãndice composto de satisfaÃ§Ã£o
df['SatisfactionComposite'] = df['JobSatisfaction'] * df['EnvironmentSatisfaction']
# Indicador OverTime
df['OverTimeBinary'] = df['OverTime'].map({'Yes':1, 'No':0})
# Risco de turnover
df['TenureRisk'] = df['TotalWorkingYears'] - df['YearsAtCompany']
# RazÃ£o de promoÃ§Ã£o
df['PromotionRatio'] = df['YearsSinceLastPromotion'] / (df['TotalWorkingYears']+ 1)
# RazÃ£o renda/idade
df['IncomeAgeRatio'] = df['MonthlyIncome'] / (df['Age'] + 1)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# ImputaÃ§Ã£o de valores faltantes
num_cols = df.select_dtypes(include=['int64','float64']).columns
imp_median = SimpleImputer(strategy='median')
df[num_cols] = imp_median.fit_transform(df[num_cols])
# Encoding categÃ³rico
cat_cols = df.select_dtypes(include=['object']).drop('Attrition',
axis=1).columns
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_df = pd.DataFrame(ohe.fit_transform(df[cat_cols]),
columns=ohe.get_feature_names_out(cat_cols))
df = pd.concat([df.drop(cat_cols, axis=1), ohe_df], axis=1)
# DefiniÃ§Ã£o de X e y
X = df.drop('Attrition', axis=1)
y = df['Attrition'].map({'Yes':1, 'No':0})
# Escalonamento e divisÃ£o
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
3
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,
stratify=y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.utils.class_weight import compute_class_weight
```

```python
# Calcular class_weights para CatBoost
classes = np.unique(y_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights_dict = dict(zip(classes, weights))

# Definindo os modelos
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000, class_weight='balanced'),
    'RandomForest': RandomForestClassifier(class_weight='balanced'),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'CatBoost': CatBoostClassifier(verbose=0, class_weights=class_weights_dict)
}

# Definindo os hiperparÃ¢metros
params = {
    'LogisticRegression': {'C':[0.01,0.1,1,10]},
    'RandomForest': {'n_estimators':[100,200], 'max_depth':[None,10,20]},
    'XGBoost': {'n_estimators':[100,200], 'learning_rate':[0.01,0.1]},
    'CatBoost': {'depth':[4,6], 'iterations':[100,200]}
}

# Treinar e buscar os melhores modelos
best_models = {}

for name, model in models.items():
    print(f"Tuning {name}...")
    grid = GridSearchCV(model, params[name], cv=5, scoring='f1', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_models[name] = grid.best_estimator_
    print(f"{name} best params: {grid.best_params_}")

# AvaliaÃ§Ã£o final dos melhores modelos
for name, model in best_models.items():
    print(f"\n=== {name} ===")
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
    proba = model.predict_proba(X_test)[:,1]
    print("AUC-ROC:", roc_auc_score(y_test, proba))
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
plt.figure(figsize=(10,6))

for name, model in best_models.items():
    proba = model.predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, proba)
    plt.plot(recall, precision, label=name)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve â€“ All Models')
plt.legend()
plt.grid()
plt.show()
```

```python
import shap
explainer = shap.TreeExplainer(best_models['RandomForest'])
shap_values = explainer.shap_values(X_train)
shap.summary_plot(shap_values, X_train, feature_names=X.columns)
```
